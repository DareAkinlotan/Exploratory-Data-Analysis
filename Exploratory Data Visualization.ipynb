## Line Charts

```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
import numpy as np
from numpy import arange
```

```python
unrate = pd.read_csv("unrate.csv")
```

```python
unrate.info()
```

```python
unrate["DATE"] = pd.to_datetime(unrate["DATE"])
```

```python
unrate.info()
```

```python
unrate.head(12)
```

```python
plt.plot()
plt.show()
```

```python
plt.plot(unrate[0:12]["DATE"], unrate[0:12]["VALUE"])
plt.xticks(rotation=90)
plt.xlabel("Month")
plt.ylabel("Unemployment Rate")
plt.title("Monthly Unemployment Trends, 1948")
plt.show()
```

## Multiple Plots

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
ax1.plot(unrate[0:11]["DATE"], unrate[0:11]["VALUE"])
plt.xticks(rotation=90)
ax2.plot(unrate[12:23]["DATE"], unrate[12:23]["VALUE"])
plt.show()
```

```python
fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
ax1.plot(unrate[0:11]["DATE"], unrate[0:11]["VALUE"])
ax1.set_title("Unemployment rate for the year 1948")
ax1.set_ylabel("unemployment rate")
ax1.set_xlabel("month")
plt.xticks(rotation=90)
ax2.plot(unrate[12:23]["DATE"], unrate[12:23]["VALUE"])
ax2.set_title("Unemployment rate for the year 1949")
ax2.set_ylabel("unemployment rate")
ax2.set_xlabel("month")
plt.show()
```

```python
unrate["MONTH"] = unrate["DATE"].dt.month
```

```python
fig = plt.figure(figsize=(6,6))
plt.plot(unrate[0:11]["MONTH"], unrate[0:11]["VALUE"], c = "red")
plt.plot(unrate[12:23]["MONTH"], unrate[12:23]["VALUE"], c = "blue")
plt.xlabel("Month")
plt.ylabel("Unemployment")
plt.show()
```

```python
fig = plt.figure(figsize=(8,6))
colors = ["red", "blue", "green", "orange", "black"]
labels = ["1948","1949","1950","1951","1952"]

for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    plt.plot(subset['MONTH'], subset['VALUE'], c=colors[i], label = labels[i])

plt.legend(loc="upper left")
plt.xlabel('Month, Integer')
plt.ylabel('Unemployment Rate, Percent')
plt.title('Monthly Unemployment Trends, 1948-1952')
plt.show()
```

## Bar and Scatter Plots

```python
reviews = pd.read_csv('fandango_scores.csv')
```

```python
reviews.info()
```

```python
reviews.head(5)
```

```python
reviews.describe()
```

```python
norm_reviews = reviews[["FILM", "RT_user_norm", "Metacritic_user_nom", "IMDB_norm", "Fandango_Ratingvalue", "Fandango_Stars"]]
norm_reviews.head(5)
```

```python
fig, ax = plt.subplots()
bar_positions = arange(5) + 0.75
num_cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']
bar_heights = norm_reviews[num_cols].iloc[0].values
ax.bar(bar_positions, bar_heights, 0.5)
tick_positions = range(1,6)
ax.set_xticks(tick_positions)
ax.set_xticklabels(num_cols, rotation=90)
ax.set_ylabel("Average Ratings")
ax.set_xlabel("Rating Source")
ax.set_title("Average User Rating For Avengers: Age of Ultron (2015)")
plt.show()
```

```python
fig, ax = plt.subplots()
bar_positions = arange(5) + 0.75
num_cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']
bar_widths = norm_reviews[num_cols].iloc[0].values
ax.barh(bar_positions, bar_widths, 0.5)
tick_positions = range(1,6)
ax.set_yticks(tick_positions)
ax.set_yticklabels(num_cols)
ax.set_xlabel("Average Ratings")
ax.set_ylabel("Rating Source")
ax.set_title("Average User Rating For Avengers: Age of Ultron (2015)")
plt.show()
```

## Scatter Plots used for correlations
`A scatter plot helps us determine if 2 columns are weakly or strongly correlated. While calculating the correlation coefficient will give us a precise number, a scatter plot helps us find outliers, gain a more intuitive sense of how spread out the data is, and compare more easily.

```python
fig, ax = plt.subplots()
ax.scatter(norm_reviews["Fandango_Ratingvalue"], norm_reviews["RT_user_norm"])
ax.set_ylabel("Rotten Tomatoes")
ax.set_xlabel("Fandango")
ax.set_title("Rotten Tomatoes Vs Fandango Ratings")
plt.show()
```

```python
fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

ax1.scatter(norm_reviews["Fandango_Ratingvalue"], norm_reviews["RT_user_norm"])
ax1.set_xlabel("Fandango")
ax1.set_ylabel("Rotten Tomatoes")
ax1.set_title("Fandango vs Rotten Tomatoes")
ax1.set_ylim(0,5)
ax1.set_xlim(0,5)

ax2.scatter(norm_reviews["Fandango_Ratingvalue"], norm_reviews["Metacritic_user_nom"])
ax2.set_xlabel("Fandango")
ax2.set_ylabel("Metacritic")
ax2.set_title("Fandango vs Metacritic")
ax2.set_ylim(0,5)
ax2.set_xlim(0,5)

ax3.scatter(norm_reviews["Fandango_Ratingvalue"], norm_reviews["IMDB_norm"])
ax3.set_xlabel("Fandango")
ax3.set_ylabel("IMDB_norm")
ax3.set_title("Fandango vs IMDB")
ax3.set_ylim(0,5)
ax3.set_xlim(0,5)

ax4.scatter(norm_reviews["Fandango_Ratingvalue"], norm_reviews["Fandango_Stars"])
ax4.set_xlabel("Fandango")
ax4.set_ylabel("Fandango_Stars")
ax4.set_title("Fandango vs F Stars")
ax4.set_ylim(0,5)
ax4.set_xlim(0,5)

plt.show()
```
