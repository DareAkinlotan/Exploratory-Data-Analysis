```python
titles = hn["title"].tolist()

python_mentions = 0
pattern = "[Pp]ython"
for s in titles:
    if re.search(pattern, s):
        python_mentions +=1
    else:
        pass
    
    
python_mentions
```

```python
titles = hn["title"].str.contains(pattern).sum()
titles
```

```python
Ruby_titles = hn[hn["title"].str.contains("[Rr]uby")]
Ruby_titles
```

```python
titles = hn["title"]
titles
```

```python
email_bool = titles.str.contains("e-?mail")
email_count = email_bool.value_counts(dropna=False)
email_count
```

```python
email_titles = titles[email_bool]
email_titles
```

```python
email_titles.size
```

```python
pattern = "\[\w+\]"
tag_titles = titles[titles.str.contains(pattern)]
tag_titles.size
```

```python
tag_titles
```

```python
pattern = r"\[(\w+)\]"
tag_tit = titles.str.extract(pattern, expand=False)
tag_tit.value_counts(dropna=False)
```

```python
def first_10_matches(pattern):
    
    all_matches = titles[titles.str.contains(pattern)]
    first_10 = all_matches.head(10)
    return first_10

first_10_matches(r"([Jj]ava)")
```

```python
Java_titles = titles[titles.str.contains(r"([Jj]ava[^Ss])")]
Java_titles.head(10).equals(first_10_matches(r"([Jj]ava[^Ss])"))
```

```python
Java_titles.head(10)
```

```python
pattern = r"\b[Jj]ava\b"
java_titles = titles[titles.str.contains(pattern)]
first_10_matches(pattern).equals(java_titles.head(10))
```
